### Experimento 6 - Woe e IV (Laplace smoothing)

Código do Experimento 5, revisado para aplicar a Suavização de Laplace, com a suavização de Lidstone quando o parâmetro α assume valores não inteiros, onde o adotou-se α = 0,5, um valor amplamente utilizado na literatura por proporcionar um bom equilíbrio entre viés e variância.

```python
def woe_iv_simples(dados, var_categorica, target_binario, alpha=0.5):
    """
    Calcula WoE e IV suavizados (Laplace smoothing) para variável categórica usando OptimalBinning.
    """
    # === 1. Preparação dos dados ===
    X = dados[[var_categorica]].astype(str)
    y = dados[target_binario]

    # === 2. Binning ótimo da variável ===
    optb = OptimalBinning(name=var_categorica, dtype="categorical")
    optb.fit(X[var_categorica].values, y.values)
    table = optb.binning_table.build()

    # === 3. Montar DataFrame base ===
    df_table = pd.DataFrame({
        "Bin": table["Bin"],
        "Count": table["Count"],
        "Event": table["Event"],
        "Non-Event": table["Non-event"],
        "Count (%)": table["Count (%)"]
    })

    # === 4. Filtrar bins válidos ===
    df_bins = df_table[
        ~df_table["Bin"].isin(["Missing", "Special"]) &
        (df_table["Count (%)"] < 1.0)
    ].copy()

    # === 5. Calcular WoE e IV suavizados ===
    total_event = df_bins["Event"].sum()
    total_nonevent = df_bins["Non-Event"].sum()
    k = len(df_bins)  # número de bins

    # Proporções suavizadas COM k
    df_bins["prop_event_smooth"] = (df_bins["Event"] + alpha) / (total_event + k * alpha)
    df_bins["prop_nonevent_smooth"] = (df_bins["Non-Event"] + alpha) / (total_nonevent + k * alpha)

    # WoE suavizado
    df_bins["WoE (smooth)"] = np.log(
        df_bins["prop_event_smooth"] / df_bins["prop_nonevent_smooth"]
    )

    # IV suavizado
    df_bins["IV (smooth)"] = (
        (df_bins["prop_event_smooth"] - df_bins["prop_nonevent_smooth"])
        * df_bins["WoE (smooth)"]
    )

    # === 6. Calcular IV total ===
    total_iv = round(df_bins["IV (smooth)"].sum(), 4)

    # === 7. Exibir resultados ===
    print(f"📊 Variable: {var_categorica}\n")
    print(df_bins[[
        "Bin", "Count", "Event", "Non-Event", "WoE (smooth)", "IV (smooth)"
    ]].round({"WoE (smooth)": 3, "IV (smooth)": 4}).to_string(index=False))
    print(f"\n🔎 Total IV (suavizado): {total_iv}\n")

    return
```
